# Pintle Test for 2.2kN Liquid Propellant Rocket Engine
Dates of testing: 1/19/19, 1/26/19 & 2/3/19

# Introduction


The Portland State Aerospace Society is in the process of testing its $2.2kN$ ($500 lb_f$) regeneratively cooled bi-propellant rocket motor. One of the critical steps prior to hot firing the motor is the testing and validating of the injector design using water.

The intitial test design intended to use turbine flow meter and an orifice plate in order to determine flow rate for each of the two propellant paths. During calibration it was found that the flow meter had been damaged and was not producing reliable data. It was decided to attempt to test the pintle using only the calibrated orifice plate and back solving for flow rates as described in the methods section.

## Underlying theory behind the experimental set up.

The first objective for testing the pintle was determining the loss coefficients for the Pintle and Annulus flow paths. For practical, cost, and safety reasons the initial testing was performed using water.

The governing equations employed:

Orifice flow coefficient for the style of orifice plate used in testing \[1\] :

\begin{equation}
C_v = \frac{Q_{gpm}}{\sqrt{\Delta P}}
\label{eq1}
\tag{1}
\end{equation}

Where;

$C_v = $ the flow coefficient 

$Q_{gpm} =$ is the volumentric flow rate in gallons per minute 

$\Delta P = $ is the difference in pressure across the orifice 


$ \\ $

Loss coefficients for Pintle and Annulus:

\begin{equation}
\dot{m} = C_LA \sqrt{2\rho(\Delta P)}
\label{eq2}
\tag{2}
\end{equation}


Where;

$C_L = $ Loss coefficient (to be calculated)

$A =$ Discharge area of Pintle/Annulus

$\dot{m} =$  Mass Flow Rate (from flow meter) 

$\Delta P$ Pressure upstream of Pintle/Annulus (gauge) - atmospheric pressure (0 psi)

$\rho = $ Density of water (calculate based on temperature)

Note: Each flow path has multiple changes to flow area between where the pressure is being measured and where it is assumed to be atmospheric. As a result, it is being assumed (possibly incorrectly) that the $C_L$ and discharge area are coupled. Given that the area does not change

## The initial test plan:

1. Calibrate pressure transducers (5 total) using a bourdon tube style pressure gauge and an air compressor.
2. With the valve to the fuel circuit closed, and the test article not attached to the system, flow water through the lox circuit at different tank pressures. For each run, obtain the flow rate using a stop watch and a 5 gallon bucket with graduated marks every 1/4 gallon.
3. Use recorded flow rate to calibrate turbine flow meter and orifice plate.
5. Attached test article to system.
6. Test LOX circuit at 50, 100, & 200 psi. Cross check flow rates from both orifice plate and turbine flow meter. Use pressure drop recorded above pinlte and flow rate data for determining loss coefficient for LOX flow path.
6. With valve to LOX circuit closed and valve to Fuel circuit open; test fuel circuit at 50, 100, & 200 psi. Use flow rate data from orifice plate and pressure recorded above annulus to determine loss coefficient for Fuel flow path.
7. Starting with Fuel circuit valve open, and LOX circuit valve closed, adjust flow through Pintle using LOX circuit glove valve unitl a 45$^{\circ}$ spray is observed. Spray angle to be measured visually using a large laser cut protractor with radial spokes every 15$^{\circ}$.  
8. With spray angle adjusted to 45$^{\circ}$, record total flow rate using orifice and LOX circuit flow rate using turbine flow meter. Fuel circuit flow rate determined via continuity. Repeat 7 and 8 for 50, 100, and 200 psi.
